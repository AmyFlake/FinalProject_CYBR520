```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data (adjust the file path as needed)
data = pd.read_csv('cybersecurity_attacks.csv')
```

```python
# Convert 'Timestamp' to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'], errors='coerce')
```

```python
# Plot 1: Time Series of Attacks
plt.figure(figsize=(10, 6))
data.set_index('Timestamp').resample('D').size().plot()
plt.title("Time Series of Attack Frequency")
plt.xlabel("Date")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
# Plot Top 10 Source IP Addresses as a Pie Chart
top_source_ips = data['Source IP Address'].value_counts().nlargest(10)

plt.figure(figsize=(8, 8))
top_source_ips.plot(kind='pie', autopct='%1.1f%%', startangle=140, cmap="Pastel1")
plt.title("Top 10 Source IP Addresses")
plt.ylabel("")  # Remove y-label for clarity
plt.show()
```
