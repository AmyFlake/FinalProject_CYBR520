```python
#1. Create a line chart to visualize the distribution of the 'capitalLong' feature.
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('cybersecurity_attacks_encoded.csv')
df.head()
```

```python
#2. Generate a bar chart to show the frequency of cybesecurity attack by 'attack type'
type_counts = df['Attack Type'].value_counts()

# Create the bar chart
plt.figure(figsize=(10, 6))
type_counts.plot(kind='bar', color=['orange', 'red', 'purple'])

# Add labels and title
plt.xlabel('Attack type')
plt.ylabel('Frequency')
plt.title('Frequency of Cybersecurity Attacks')

#  Display the graph on the screen. 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('cybersecurity_attacks_transformed_cleaned.csv')

# Ensure the time column is in datetime format (replace 'Time Column' with the actual column name)
df['Timestamp'] = pd.to_datetime(df['Timestamp'])  # Replace 'Time' with your timestamp column

# Group data by year and count frequencies of attack types
df['Year'] = df['Timestamp'].dt.year
attack_counts = df.groupby('Year')[['Attack Type_DDoS', 'Attack Type_Malware', 'Attack Type_Intrusion']].sum()

# Plot a stacked histogram
attack_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['skyblue', 'orange', 'green'])

# Customize the chart
plt.title('Frequency of Attack Types Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(title='Attack Types')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#5. Create a pie chart to represent the proportion of attack types.
attack_counts = df['Attack Type'].value_counts()

# Create the pie chart
plt.figure(figsize=(10, 6))
plt.pie(
    attack_counts, 
    labels=attack_counts.index, 
    autopct='%1.0f%%', 
    startangle=90
)
plt.title('Attack Types')
plt.axis('equal')  # Ensures the pie chart is circular
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('cybersecurity_attacks_transformed_cleaned.csv')

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Set the diagonal (self-correlations) to NaN
np.fill_diagonal(correlation_matrix.values, np.nan)

# Flatten the matrix into a Series
correlation_pairs = correlation_matrix.unstack()

# Find the highest positive correlation
highest_positive = correlation_pairs.idxmax()
highest_positive_value = correlation_pairs.max()

# Find the highest negative correlation
highest_negative = correlation_pairs.idxmin()
highest_negative_value = correlation_pairs.min()

# Display the results
print(f"Highest Positive Correlation: {highest_positive} with a value of {highest_positive_value:.2f}")
print(f"Highest Negative Correlation: {highest_negative} with a value of {highest_negative_value:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('cybersecurity_attacks_transformed_cleaned.csv')

# List of attack type columns
attack_columns = ['Attack Type_DDoS', 'Attack Type_Malware', 'Attack Type_Intrusion']

# Group by 'Geo-location Data' and calculate the sum of attack counts for each type
attack_counts_by_region = df.groupby('Geo-location Data')[attack_columns].sum()

# Plot the result as a line chart
plt.figure(figsize=(12, 6))

for attack_type in attack_columns:
    plt.plot(attack_counts_by_region.index, attack_counts_by_region[attack_type], label=attack_type)

plt.title('Frequency of Attack Types by Region (Line Chart)')
plt.xlabel('Region (Geo-location Data)')
plt.ylabel('Attack Count')
plt.legend(title='Attack Types')
plt.grid(visible=True)
plt.tight_layout()
plt.show()
```
